```python
import pandas as pd
from permute.core import one_sample
from permute.core import two_sample
import numpy as np
from numpy.random import RandomState
from mlxtend.evaluate import permutation_test
from netneurotools import stats as nnstats
```

```python
dfba = pd.read_csv("ba-scoresPerTopic-final.tsv", sep="\t")
dfct = pd.read_csv("ct-scoresPerTopic-final.tsv", sep="\t").fillna(0)
dfct.loc[:,"201801":"201940"]
```

```python
bareslist = []
prng = RandomState(42)
for i in range(len(dfba)):
    data = dfba.loc[i,"201701":"201940"]
    ref_data = dfba.loc[0,"201701":"201940"]
    pval = nnstats.permtest_rel(ref_data, data,n_perm=20000)
    # Working alternative to the nnstats implementation but returns 0.0 for the reference data with itself which should result to 1.0.
    # However, besides this, the results seem to be virtually the same like with the nnstats package and one-tailed tests are possible. Keep it around.
    #pval, _ = one_sample(ref_data, data,plus1=False, alternative='two-sided', stat='mean', seed=prng)
    groupname = dfba.loc[i]["ablationgroup"]
    
    bareslist.append((groupname, pval))
    print("[ba]",groupname,pval)
bareslist
```
