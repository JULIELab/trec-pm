```python
eval_df = pd.read_csv("mydevset_triples.tsv", sep="\t", names=["text_a", "text_b", "labels"])
#eval_df = pd.read_csv("eval_question_text.tsv", sep="\t", names=["text_a", "text_b", "labels"])
#eval_df = pd.read_csv("eval_question_title.tsv", sep="\t", names=["text_a", "text_b", "labels"])
#eval_df = pd.read_csv("eval_question_abstract.tsv", sep="\t", names=["text_a", "text_b", "labels"])
```

```python
model_args = ClassificationArgs()
model_args.eval_batch_size = 16
model_args.dataloader_num_workers = 1
model_args.manual_seed = 4
model_args.dataloader_num_workers = 1
model_args.labels_list = [0, 1]
model_args.max_seq_length = 512
```

```python
model = ClassificationModel(
    "roberta",
    "outputs/best_model",
    args=model_args,
    cuda_device=1
)
```

```python
input_lists = [list(l) for l in eval_df[["text_a", "text_b"]].values]
predictions, probs = model.predict(input_lists)
```

```python
sklearn.metrics.accuracy_score(eval_df["labels"].values, predictions)
```
